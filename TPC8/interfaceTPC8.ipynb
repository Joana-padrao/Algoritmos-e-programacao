```python
import PySimpleGUI as sg

def getEMD(linha):
    emd=[]
    novalinha=linha.strip("\n")
    campos=novalinha.split(",")
    emd.append("emd"+str(int(campos[1])+1))
    for i in range (2, len(campos)):
        emd.append(campos[i])
    return emd

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f:
        emd=getEMD(linha)
        bd.append(emd)
    return bd

BD = []
BD = lerDataset("emd.csv")


def modalidades(bd):
    l=[]
    for emd in bd:
        if (emd[7] not in l):
            l.append(emd[7])
    l.sort()
    return l

def chaveOrdem(exame):
    l= exame[1].split("-")
    return l[0]
    

def distribPorAno(bd):
    bd.sort(key=chaveOrdem, reverse=True)
    distribuicao={}
    for emd in bd:
        li=emd[1].split("-")
        if li[0] in distribuicao.keys():
            distribuicao[li[0]]=distribuicao[li[0]] +1
        else:
            distribuicao[li[0]]=1   

    return distribuicao

def distribPorClube(bd):
    distribuicao={}
    for emd in bd:
        if emd[8] in distribuicao.keys():
            distribuicao[emd[8]]=distribuicao[emd[8]] +1
        else:
            distribuicao[emd[8]]=1   

    return distribuicao

def chaveOrdem(exame):
    return exame[1] 
    
def listarDataset(bd):
    LIS=[]
    bd.sort(key=chaveOrdem, reverse=True)
    LIS.append(("id   data   nome                  apto"))
    LIS.append(("------------------------"))
    for emd in bd:
        LIS.append((emd[0] + "|" + emd[1] + "|" + emd[3] + " " + emd[2] + "|" + emd[11]))
    return(LIS)
        
        
        
tarefas=[]

interfacePrincipal = [[sg.Button("Carregar BD"), sg.Button("modalidades"), sg.Button("distribuicao ano"), 
                       sg.Button("clube"), sg.Button("listar")]]

data_viewer_column = [
    [sg.Text("Painel de Dados")],
    [sg.Text(size=(40, 10), key="-Dados-")],
    [sg.Listbox(values=tarefas, size=(40,10), key="_tarefas")]
]

layout = [
    [
        sg.Column(interfacePrincipal),
        sg.VSeperator(),
        sg.Column(data_viewer_column),
    ]
]

window = sg.Window("emd", layout)

stop = False
# Create an event loop

while not stop:
    event, values = window.read()
    # End program if user closes window or
    # presses the OK button
    if event == sg.WIN_CLOSED:
        stop = True
    elif event == "Carregar BD":
        window["_tarefas"].Update("A BD tem " + str(len(BD))+ " elementos")
    elif event == "modalidades":
        window["_tarefas"].Update(str(modalidades(BD)))
    elif event == "distribuicao ano":
        window["_tarefas"].Update(str(distribPorAno(BD)))
    elif event == "clube":
        window["_tarefas"].Update(str(distribPorClube(BD)))
    elif event == "listar":
        window["_tarefas"].Update(str(listarDataset(BD)))
            
    else:
        print (event + ":\n")
        print (values)


window.close()
```
