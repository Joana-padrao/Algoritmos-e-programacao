## Leitura (exercÃ­cio 1)

Depois de pensares num modelo adequado especifica a funÃ§Ã£o que lÃª o ficheiro e povoa esse modelo.

Nota: os valores do campo id sÃ£o demasiado complexos para serem usados na aplicaÃ§Ã£o pelo utilizador.

Ideia: Cria ids novos, mais simples, tipo: emd00001, emd00002, ..., emd00100, ...

````python
# Descreve em texto o teu 
# 
# BdEMD = [EMD]
# EMD = [id, dataEMD, pnome, unome, idade, genero, morada, moralidade, clube, email, federado, resultado]
#
# Leitura/carregamento da informaÃ§Ã£o do ficheiro

def getEMD(linha):
    emd=[]
    novalinha=linha.strip("\n")
    campos=novalinha.split(",")
    emd.append("emd"+str(int(campos[1])+1))
    for i in range (2, len(campos)):
        emd.append(campos[i])
    return emd

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f:
        emd=getEMD(linha)
        bd.append(emd)
    return bd
````

````python
BD = []
BD = lerDataset("emd.csv")
print(BD)
````

## Modalidades (exercÃ­cio 4)

Especifica um funÃ§Ã£o que dÃ¡ como resultado uma lista de modalidades ordenada alfabeticamente e sem repetiÃ§Ãµes.

````python
def modalidades(bd):
    l=[]
    for emd in bd:
        if (emd[7] not in l):
            l.append(emd[7])
    l.sort()
    return l 
print(modalidades(BD))
````

## DistribuiÃ§Ã£o por modalidade (exercÃ­cio 5)

````python
def distribPorModalidade(bd):
    distribuicao={}
    for emd in bd:
        if emd[7] in distribuicao.keys():
            distribuicao[emd[7]]=distribuicao[emd[7]] +1
        else:
            distribuicao[emd[7]]=1   

    return distribuicao
print(distribPorModalidade(BD))
````

## DistribuiÃ§Ã£o por ano (exercÃ­cio 7)

Especifica uma funÃ§Ã£o que dÃ¡ como resultado uma lista de pares indicando quantos EMD estÃ£o registados por cada ano.

````python
def chaveOrdem(exame):
    l= exame[1].split("-")
    return l[0]
    

def distribPorAno(bd):
    bd.sort(key=chaveOrdem, reverse=True)
    distribuicao={}
    for emd in bd:
        li=emd[1].split("-")
        if li[0] in distribuicao.keys():
            distribuicao[li[0]]=distribuicao[li[0]] +1
        else:
            distribuicao[li[0]]=1   

    return distribuicao
print(distribPorAno(BD))
````

## DistribuiÃ§Ã£o parametrizada (exercÃ­cio 8)

````python
def invalida():
    print("Opção inválida")

def distribPorIdade(bd):
    distribuicao={}
    for emd in bd:
        if emd[4] in distribuicao.keys():
            distribuicao[emd[4]]=distribuicao[emd[4]] +1
        else:
            distribuicao[emd[4]]=1  
    return distribuicao

def distribPorGenero(bd):
    distribuicao={}
    for emd in bd:
        if emd[5] in distribuicao.keys():
            distribuicao[emd[5]]=distribuicao[emd[5]] +1
        else:
            distribuicao[emd[5]]=1 
    return distribuicao

def distribPorMorada(bd):
    distribuicao={}
    for emd in bd:
        if emd[6] in distribuicao.keys():
            distribuicao[emd[6]]=distribuicao[emd[6]] +1
        else:
            distribuicao[emd[6]]=1  
    return distribuicao 

def distribPorFederado(bd):
    distribuicao={}
    for emd in bd:
        if emd[10] in distribuicao.keys():
            distribuicao[emd[10]]=distribuicao[emd[10]] +1
        else:
            distribuicao[emd[10]]=1 
    return distribuicao

def distribPorResultado(bd):
    distribuicao={}
    for emd in bd:
        if emd[11] in distribuicao.keys():
            distribuicao[emd[11]]=distribuicao[emd[11]] +1
        else:
            distribuicao[emd[11]]=1   
    return distribuicao

lista=[]
def distrib(bd):
    op="1"
    while op!="0":
        op=input('''
           Deseja distribuir por
          (1) Ano
          (2) Idade
          (3) Género
          (4) Morada
          (5) Modalidade
          (6) Clube
          (7) Federados
          (8) Resultado

          
          Qual a sua opção? ''')
          
        lista.append(op)
        if (op=="1"):
            print(distribPorAno(BD))

        elif (op=="2"):
            print(distribPorIdade(BD))

        elif (op=="3"):
            print(distribPorGenero(BD))
        
        elif (op=="4"):
            print(distribPorMorada(BD))

        elif (op=="5"):
            print(distribPorModalidade(BD))

        elif (op=="6"):
            print(distribPorClube(BD))
        
        elif (op=="7"):
            print(distribPorFederado(BD))

        elif (op=="8"):
            print(distribPorResultado(BD))

        else:
            invalida()

distrib(BD)

````
